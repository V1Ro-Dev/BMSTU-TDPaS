```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock_data = pd.read_csv('stock_data.csv')
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Price'], label='Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Price Over Time')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
window_size = 10  # Задайте размер окна для скользящего среднего
stock_data['Moving_Average'] = stock_data['Price'].rolling(window=window_size).mean()

plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Price'], label='Price', color='blue')
plt.plot(stock_data['Date'], stock_data['Moving_Average'], label=f'{window_size}-Day Moving Average', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Price with Moving Average')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
