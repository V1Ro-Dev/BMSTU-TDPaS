```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('web_traffic_data.csv')
```

```python
start_date = pd.to_datetime("2024-12-24")
data['Date'] = start_date + pd.to_timedelta(data['Day'], unit='D')

plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Page_Views'],color='blue')
plt.xlabel('Date')
plt.ylabel('Page Views')
plt.title('Page Views Over Time')
plt.grid(True)
plt.show()
```

```python
data['Month'] = data['Date'].dt.month
monthly_stats = data.groupby('Month')['Page_Views'].agg(['mean', 'median']).reset_index()
monthly_stats.rename(columns={'mean': 'Average Page Views', 'median': 'Median Page Views'}, inplace=True)

print("      :")
print(monthly_stats)
```
